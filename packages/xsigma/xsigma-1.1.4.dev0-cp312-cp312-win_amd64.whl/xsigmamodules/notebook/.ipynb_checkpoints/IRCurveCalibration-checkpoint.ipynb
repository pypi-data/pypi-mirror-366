```python
marketContainer = anyContainer()

marketContainer.insert(anyId(calendarId(calendar_key)), anyObject(holiday_list))

marketContainer.insert(
    anyId(valuationDatetimeId()), anyObject(valuationDatetime(valuation_date))
)

marketContainer.insert(
    anyId(swapDefaultConventionId(ccy)), anyObject(swapDefaultConfig)
)

marketContainer.insert(
    anyId(futureDefaultConventionId(ccy, index_name, "1m")), anyObject(idxData[0])
)

marketContainer.insert(
    anyId(futureDefaultConventionId(ccy, index_name, "3m")), anyObject(idxData[1])
)
```

```python
custom_config = curveCalibrationConfig(
    2.0,  # bootstrap_stdev
    0.005,  # smoothing_weight
    1.0e-8,  # ceres_function_tolerance
    1.0e-8,  # ceres_gradient_tolerance
    1.0e-8,  # ceres_parameter_tolerance
    -3.0,  # parameter_lower_bound
    3.0,  # parameter_upper_bound
    6.0e-02,  # lm_lambda
    0.2,  # lm_epsilon
    1.0e-8,  # lm_function_tolerance
    500,  # max_iterations
    False,  # use_ceres
    dates_mode_type.INSTRUMENT,  # MERGED REPLACED INSTRUMENT
    interpolation_enum.LINEAR,
    interpolation_enum.CUBIC_SPLINE,
    False,
)
```

```python
obj = irCurveCalibration(
    valuation_date,
    ccy,
    index_name,
    discount_def_id,
    calendar_key,
    targetOptions,
    marketContainer,
    custom_config,
)

obj.calibrate(False)
obj.debug()

obj.calibrate(True)
obj.debug()
```

```python
market = obj.market()
```

```python
id = discountCurveId(ccy, discount_def_id)
```

```python
curve = obj.curve(id)
```

```python
idf = forecastCurveId(ccy, index_name, "1b")
dates = obj.node_dates(idf)
```

```python
print(dates)
```

```python
curve.df(valuation_date)
```

```python
import numpy as np

dfs = [curve.df(date) for date in dates]
t = [(date - valuation_date + 1) / 365.0 for date in dates]

dfs = -np.log(dfs) / t
dfs[0] = dfs[1]
t_rates = t
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(
    t_rates, dfs, marker="o", linestyle="-", color="b", label="Discount Factor Curve"
)
plt.xlabel("Time to Maturity (Years)")
plt.ylabel("Discount Factor")
plt.title("Discount Factor Curve (Continuous Compounding)")
# plt.xlim(0, 3)
plt.grid(True)
plt.legend()
plt.show()
```
